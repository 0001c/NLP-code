# 获取词典和最大长度

```python
def read_word_dict(file_path):

    """
    读取词典文件，返回词典和最大长度
    :param file_path: 词典文件路径
    :return: 词典和最大长度
    """

    dictionary = set()
    max_len = 0


    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f.readlines():
            line = line.strip()
            if line:
                dictionary.add(line)
                line_len = len(line)
                max_len = max(max_len, line_len)
    return dictionary, max_len
```

```python
dictionary, max_len = read_word_dict('data/word_dict.txt')
```

# 正向最大匹配法

```python
def forward_max_match(text, dictionary, max_len):
    """
    正向最大匹配法
    :param text: 输入文本
    :param dictionary: 词典
    :param max_len: 最大匹配长度
    :return: 分词结果
    """

    result = []
    head = 0
    while head < len(text):
        for j in range(min(head + max_len, len(text)), head, -1):
            if text[head:j] in dictionary:
                result.append(text[head:j])
                head = j
                break
        else:
            result.append(text[head])
            head += 1
    return result
```

```python
text = '南京市长江大桥'
text_cut = forward_max_match(text, dictionary, max_len)
print(text_cut)
```

```python
def backward_max_match(text, dictionary, max_len):
    """
    逆向最大匹配法
    :param text: 输入文本
    :param dictionary: 词典
    :param max_len: 最大匹配长度
    :return: 分词结果
    """
    result = []
    tail = len(text) - 1
    while tail >= 0:
        for j in range(max(tail - max_len, -1), tail):
            if text[j:tail+1] in dictionary:
                result.append(text[j:tail+1])
                tail = j - 1
                break
        else:
            result.append(text[tail])
            tail -= 1
    return result
```

```python
text = '南京市长江大桥'
text_cut = forward_max_match(text, dictionary, max_len)
print("正向最大匹配法：")
print(text_cut)
text_cut = backward_max_match(text, dictionary, max_len)
print("反向最大匹配法：")
print(text_cut[::-1])
```

# 双向最大匹配法

```python
def bidirectional_max_match(text, dictionary, max_len):
    """
    双向最大匹配法
    """
    fmm = forward_max_match(text, dictionary, max_len)
    bmm = backward_max_match(text, dictionary, max_len)
    bmm = bmm[::-1]
    return fmm if len(fmm) < len(bmm) else bmm
```

```python
text = '南京市长江大桥'
text_cut = bidirectional_max_match(text, dictionary, max_len)
print("双向最大匹配法：")
print(text_cut)
```
