# pkuseg分词
多领域中文分词工具包


## 特点 
1. 支持多领域分词，如医疗、法律、金融等。
2. 支持自定义词典，用户可以根据自己的需求添加或删除词。
3. 支持批量处理，用户可以对多文本进行分词。
3. 更高的分词准确率，比jieba更准确。
4. 支持用户自训练模型，用户可以根据自己的需求训练模型，提高分词准确率。
5. 支持词性标注，用户可以根据词性进行分词。


## 主要功能
### 初始化分词器
```python

pkuseg.pkuseg(model = "default", user_dict = None, postag = True)  
'''
model_name 模型路径  
 default：默认模型，包含中文、英文、数字、标点符号等。  
 news：新闻领域模型，包含新闻相关词汇。  
 web：网络领域模型，包含网络相关词汇。  
 medical：医疗领域模型，包含医疗相关词汇。  
 tourism：旅游领域模型，包含旅游相关词汇。  
 model_path：自定义模型路径.  
user_dict：用户词典路径。  
 default：默认词典，包含中文、英文、数字、标点符号等。  
 None：不使用用户词典。  
 user_dict_path：自定义词典路径。可以传入列表。  
postag：是否进行词性标注。  
 True：进行词性标注。  
 False：不进行词性标注。  
'''
```


####  加载安装包

````python
import pkuseg
````

> ##### pip 安装 pkuseg 报错 ModuleNotFoundError: No module named 'numpy' 的原因与解决
> 原因：pkuseg 的 setup.py 在构建时直接 import numpy，但其 pyproject.toml 中未声明 numpy 为构建依赖。pip 默认启用构建隔离（build isolation），构建子进程无法继承当前环境已安装的 numpy，导致失败。

> 解决方案：
> ```bash
> pip install pkuseg --no-build-isolation
> ```
> 该参数禁用构建隔离，直接使用当前环境（需确保 numpy 已预先安装）。
> 本质：这是 pkuseg 项目的缺陷（年久失修，未适配现代打包规范），非用户环境问题。


#### 默认初始化模型

````python
seg = pkuseg.pkuseg()
````

````python
seg_list = seg.cut("李小福是创新办主任也是云计算方面的专家")
print(seg_list)
````

#### 自定义词典

````python
seg = pkuseg.pkuseg()
````

````python
seg_list = seg.cut("解决解放大道路面积水")
print(seg_list)
````

````python
userdict = [line.strip() for line in open("data/userdict/userdict.txt", "r", encoding="utf-8")]
print(userdict)
````

````python
seg = pkuseg.pkuseg(user_dict=userdict)
seg_list = seg.cut("解决解放大道路面积水")
print(seg_list)
````

#### 进行词性标注

````python
seg = pkuseg.pkuseg(postag = True)
seg_list = seg.cut("李小福是创新办主任也是云计算方面的专家")
print(seg_list)
````
