# jieba 分词

```python
import jieba
seg_list = jieba.cut("我来到北京清华大学",cut_all=True)
print("全模式：" + " ".join(seg_list))

seg_list = jieba.cut("我来到北京清华大学",cut_all=False)
print("精确模式：" + " ".join(seg_list))

seg_list = jieba.cut("我来到北京清华大学")
print("默认模式：" + " ".join(seg_list))  # 默认模式也是精准模式

seg_list = jieba.cut_for_search("小明硕士毕业于中国科学院计算所，后在日本京都大学深造")
print("搜索引擎模式：" + " ".join(seg_list))
```

## 去除停用词
    停用词：一些常见的词，对文本的分析没有帮助，例如“的”、“是”、“和”等。

```python
# 读取停用词
stopwords = [line.strip() for line in open("data/stopwords/stopwords.txt", "r", encoding="utf-8").readlines()]
print(stopwords)
```

```python
seg_list = jieba.cut("李小福是创新办主任也是云计算方面的专家")
# 去除停用词
seg_list = [word for word in seg_list if word not in stopwords]
print(" ".join(seg_list))
```

## 词性标注
POS：Part-of-Speech的缩写，即词性。

```python
import jieba.posseg as pseg
words = pseg.cut("我爱北京天安门")
print(words)
# 提取词和词性
for word in words:
    print(word.word, word.flag)

```

### 停用词过滤

```python
stopwords = 'data/stopwords/stopwords_phone.txt'
with open(stopwords, 'r', encoding='utf-8') as f:
    stopwords = f.readlines()
stopwords = [line.strip() for line in stopwords]
seg_list = jieba.cut(demo_txt)
print(','.join([word for word in list(seg_list) if word not in stopwords and word != ' ']))

```

### 自定义词典
