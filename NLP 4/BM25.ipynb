# BM25

```python
import jieba
from rank_bm25 import BM25Okapi
```

#### 自定义词典

```python
jieba.load_userdict('data/userdict_mb.txt')
```

#### 停用词

```python
stopwords = [line.strip() for line in open('data/stopwords.txt', 'r', encoding='utf-8').readlines()]
```

#### 读取文件，分词

```python
def readfile2wordlist(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    wordlist = []
    cut_wordlist = []
    for line in lines:
        words = jieba.cut(line.strip())
        words = [word for word in words if word not in stopwords and word != ' ']
        wordlist.append(line.strip())
        cut_wordlist.append(words)
    return wordlist, cut_wordlist
```

#### 构建BM25语料库，计算BM25值

```python
def calculate_bm25(fliepath):
    wordlist, cut_wordlist = readfile2wordlist(fliepath)
    bm25 = BM25Okapi(cut_wordlist)
    return wordlist, bm25
```

#### 针对输入值分词

```python
def split_text(text):
    words = jieba.cut(text)
    words = [word for word in words if word not in stopwords and word != ' ']
    return words
```

#### 计算相似度

```python
def calcualte_similarity(input_text, wordlist, bm25):
    input_words = split_text(input_text)
    scores = bm25.get_scores(input_words)
    sorted_scores = sorted(enumerate(scores), key=lambda x: x[1], reverse=True)
    top_scores = sorted_scores[:5]
    results = [(wordlist[i], score) for i, score in top_scores]
    return results
```

```python
wordlist,bm25 = calculate_bm25('data/mb.txt')
```

```python
word = "Apple iPhone 7 （A1660） 128G 玫瑰金色 移动联通电信4G手机"
results = calcualte_similarity(word, wordlist, bm25)
for result in results:
    print(f"Text: {result[0]}, Similarity Score: {result[1]}")
```
